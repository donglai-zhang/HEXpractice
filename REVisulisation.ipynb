```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# true data
df = pd.read_csv("../../py_data/HEXPractice/lumpHEX/rinlet/parallel.csv", header=0)
x = df["Day"]
```

```python
# cases:
d_path = Path("../../py_data/HEXPractice/RN")
cno = 2

dfl = pd.read_csv(f"{d_path}/case{cno}/preds_l.csv", header=0)
dfm = pd.read_csv(f"{d_path}/case{cno}/preds_m.csv", header=0)
dfh = pd.read_csv(f"{d_path}/case{cno}/preds_h.csv", header=0)
```

```python
plt.figure(figsize=(15, 3))
plt.plot(dfh["F1i"], c="red", label="Cold (20% noises)")
plt.plot(dfm["F1i"], c="orange", label="Cold (10% noises)")
plt.plot(dfl["F1i"], c="green", label="Cold (5% noises)")
plt.plot(df["F1i"], c="black", alpha=0.8, label="Cold (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Temperature (K)")
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(dfh["F2i"], c="red", label="Hot (20% noises)")
plt.plot(dfm["F2i"], c="orange", label="Hot (10% noises)")
plt.plot(dfl["F2i"], c="green", label="Hot (5% noises)")
plt.plot(df["F2i"], c="black", alpha=0.8, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Temperature (K)")
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(dfh["F1m"], c="red", label="Cold (20% noises)")
plt.plot(dfm["F1m"], c="orange", label="Cold (10% noises)")
plt.plot(dfl["F1m"], c="green", label="Cold (5% noises)")
plt.plot(df["F1m"], c="black", alpha=0.8, label="Cold (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Mass flow (kg/s)")
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(dfh["F2m"], c="red", label="Hot (20% noises)")
plt.plot(dfm["F2m"], c="orange", label="Hot (10% noises)")
plt.plot(dfl["F2m"], c="green", label="Hot (5% noises)")
plt.plot(df["F2m"], c="black", alpha=0.8, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Mass flow (kg/s)")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
fig.set_figheight(10)
fig.set_figwidth(6)
ax[0].plot(dfl["Sigma1_p"], c="green", label="Pred (5% noises)")
ax[0].plot(df["Sigma1"], c="black", alpha=0.8, label="True (no noise)")
ax[0].legend(loc="upper left")
ax[0].set_ylabel("Fouling thickness (m)")
ax[0].set_xlim(1, 200)
ax[1].plot(dfl["k_p"], c="green", label="Pred (5% noises)")
ax[1].plot(0.2 * np.ones(len(x)), c="black", alpha=0.8, label="True (no noise)")
ax[1].legend(loc="upper left")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Deposit conductivity (K/W)")
ax[1].set_xlim(1, 200)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
fig.set_figheight(10)
fig.set_figwidth(6)
ax[0].plot(dfm["Sigma1_p"], c="orange", label="Pred (10% noises)")
ax[0].plot(df["Sigma1"], c="black", alpha=0.8, label="True (no noise)")
ax[0].legend(loc="upper left")
ax[0].set_ylabel("Fouling thickness (m)")
ax[1].plot(dfm["k_p"], c="orange", label="Pred (10% noises)")
ax[1].plot(0.2 * np.ones(len(x)), c="black", alpha=0.8, label="True (no noise)")
ax[1].legend(loc="lower left")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Deposit conductivity (K/W)")
ax[0].set_xlim(1, 200)
ax[1].set_xlim(1, 200)
plt.show()
```
