```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
```

### Complete noise

```python
plt.figure(figsize=(15, 6))
plt.plot(dff["F1i"], c="blue", label="Cold (complete noises)")
plt.plot(df["F1i"], c="black", alpha=0.5, label="Cold (no noise)")

plt.plot(dff["F2i"], c="red", label="Hot (complete noises)")
plt.plot(df["F2i"], c="grey", alpha=1, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Temperature (K)")
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(dff["F1i"] - df["F1i"], c="blue", label="Cold (complete noises)")
plt.plot(dff["F2i"] - df["F2i"], c="red", label="Hot (complete noises)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Temperature difference (K)")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(dff["F1m"], c="blue", label="Cold (complete noises)")
plt.plot(df["F1m"], c="black", alpha=0.5, label="Cold (no noise)")

plt.plot(dff["F2m"], c="red", label="Hot (complete noises)")
plt.plot(df["F2m"], c="grey", alpha=1, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Mass flow rate (kg/s)")
plt.show()
```

```python
Q_true = df["Q"] 
Q_pf1 = dff["F1m"] * 2916 * (dff["F1o"] - dff["F1i"])
Q_pf2 = dff["F2m"] * 4180 * (dff["F2i"] - dff["F2o"])


dT1 = df["F2i"] - df["F1i"]
dT2 = df["F2o"] - df["F1o"]
dTlm = (dT1 - dT2) / np.log(dT1 / dT2)
UA = Q_true / dTlm

dT1p = dff["F2i"] - dff["F1i"]
dT2p = dff["F2o"] - dff["F1o"]
dTlmp = (dT1p - dT2p) / np.log(dT1p / dT2p)
UAp = Q_pf1 / dTlmp

fig, ax = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(15)
ax.plot(Q_pf1, c="blue", label="Noisy $Q$ by cold")
ax.plot(Q_pf2, c="red", label="Noisy $Q$ by hot")
ax.plot(Q_true, c="black", label="True $Q$")
ax.legend(loc="upper right")
ax.set_xlabel("Days")
ax.set_ylabel("Heat duty (J/s)")
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(3)
fig.set_figwidth(15)
ax.plot((Q_pf1 - Q_true) / Q_true * 100, c="blue", label="Noisy $Q$ by cold")
ax.plot((Q_pf2- Q_true) / Q_true * 100, c="red", label="Noisy $Q$ by hot")
ax.legend(loc="upper right")
ax.set_xlabel("Days")
ax.set_ylabel("Heat duty relative error %")
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_figheight(3)
fig.set_figwidth(15)
ax.plot((Q_pf1 - Q_true), c="blue", label="Noisy $Q$ by cold")
ax.plot((Q_pf2- Q_true), c="red", label="Noisy $Q$ by hot")
ax.legend(loc="upper right")
ax.set_xlabel("Days")
ax.set_ylabel("Heat duty absolute error (J/s)")
plt.show()
```
