```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# true data
df = pd.read_csv("../../py_data/HEXPractice/lumpHEX/rinlet/parallel.csv", header=0)
x = df["Day"]
```

```python
# cases:
d_path = Path("../../py_data/HEXPractice/RN")
cno = 1

dfl = pd.read_csv(f"{d_path}/case{cno}/preds_l.csv", header=0)
dfm = pd.read_csv(f"{d_path}/case{cno}/preds_m.csv", header=0)
dfh = pd.read_csv(f"{d_path}/case{cno}/preds_h.csv", header=0)
dff = pd.read_csv(f"{d_path}/case{cno}/preds_full.csv", header=0)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(dff["F1i"], c="blue", label="Cold (complete noises)")
plt.plot(df["F1i"], c="black", alpha=0.5, label="Cold (no noise)")

plt.plot(dff["F2i"], c="red", label="Hot (complete noises)")
plt.plot(df["F2i"], c="grey", alpha=1, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Temperature (K)")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(dff["F1m"], c="blue", label="Cold (complete noises)")
plt.plot(df["F1m"], c="black", alpha=0.5, label="Cold (no noise)")

plt.plot(dff["F2m"], c="red", label="Hot (complete noises)")
plt.plot(df["F2m"], c="grey", alpha=1, label="Hot (no noise)")
plt.legend(loc="upper right")
plt.xlabel("Days")
plt.ylabel("Mass flow rate (kg/s)")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_figheight(3)
fig.set_figwidth(15)
ax[0].plot(dff["Sigma1_p"], c="purple", label="Pred (complete noises)")
ax[0].plot(df["Sigma1"], c="black", alpha=0.8, label="True (no noise)")
ax[0].legend(loc="upper left")
ax[0].set_xlabel("Days")
ax[0].set_ylabel("Fouling thickness (m)")
ax[1].plot(dff["k_p"], c="purple", label="Pred (complete noises)")
ax[1].plot(0.2 * np.ones(len(x)), c="black", alpha=0.8, label="True (no noise)")
ax[1].legend(loc="lower left")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Deposit conductivity (K/W)")
ax[0].set_xlim(1, 200)
ax[1].set_xlim(1, 200)
ax[1].set_ylim(-5, 5)
plt.show()
```

```python
Q_true = df["F1m"] * 680 * (df["F1o"] - df["F1i"])
Q_pf = dff["F1m"] * 680 * (dff["F1o"] - dff["F1i"])

dT1 = df["F2i"] - df["F1i"]
dT2 = df["F2o"] - df["F1o"]
dTlm = (dT1 - dT2) / np.log(dT1 / dT2)
UA = Q_true / dTlm

dT1p = dff["F2i"] - dff["F1i"]
dT2p = dff["F2o"] - dff["F1o"]
dTlmp = (dT1p - dT2p) / np.log(dT1p / dT2p)
UAp = Q_pf / dTlmp
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_figheight(3)
fig.set_figwidth(15)
ax[0].plot(Q_p1, c="purple", label="Noisy $Q$" )
ax[0].plot(Q_true, c="black", label="True $Q$")
ax[0].legend(loc="upper right")
ax[0].set_xlabel("Days")
ax[0].set_ylabel("Heat duty (K/s)")
ax[0].set_xlim(1, 200)

ax[1].plot(1 / UAp, c="purple", label="Noisy $R$" )
ax[1].plot(1 / UA, c="black", label="True $R$")
ax[1].legend(loc="upper right")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Overall thermal resistance (K/W)")
ax[1].set_xlim(1, 200)

plt.show()
```
